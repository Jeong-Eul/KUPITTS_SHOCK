```python
import pandas as pd
from tqdm.notebook import tqdm
import re
```

```python
feature_ID_list = pd.read_csv('./Feature ID list for shock(MIMIC-4).csv')
feature_name = feature_ID_list['Feature name'].tolist()
feature_ID_list.set_index('Feature name', inplace=True)
```

```python
data = pd.read_csv('./pivoted lab table.csv')
data = data.dropna(subset=['hadm_id'])
data.drop(columns=['stay_id'], inplace=True)
patients = pd.read_csv('./icustays.csv')
patients.drop(columns=['stay_id'], inplace=True)
#patients.drop(columns=['Unnamed: 0'], inplace=True)
#patients.drop(columns=['ICUSTAY_ID'], inplace=True)

data = pd.merge(data, patients, on=['subject_id','hadm_id'], how='left')
#data = data[data['los']>1]
data.reset_index(drop=True, inplace=True)
```

```python
data['charttime'] = pd.to_datetime(data['charttime'])
data['outtime'] = pd.to_datetime(data['outtime'])
data['intime'] = pd.to_datetime(data['intime'])

condition1 = (data['intime'] > data['charttime'])
condition2 = (data['intime'] - pd.to_timedelta(24, unit='h') < data['charttime'])

before_icu_data = data[condition1 & condition2]
before_icu_data = before_icu_data.sort_values(by=['subject_id', 'hadm_id', 'charttime'], ascending=[True,True,True])
before_icu_data.reset_index(drop=True, inplace=True)

before_icu_data.fillna(method='ffill', inplace=True)

def latest_data(x):
    x.reset_index(drop=True, inplace=True)
    df = x.loc[len(x)-1]

    return df

before_icu_data_latest = before_icu_data.groupby('hadm_id').apply(latest_data)
before_icu_data_latest.reset_index(drop=True, inplace=True)
```

```python
bidl_ls = before_icu_data_latest['hadm_id'].tolist()

data = pd.read_csv('./pivoted chart table.csv')
data = data.dropna(subset=['hadm_id'])
patients = pd.read_csv('./icustays.csv')
#patients.drop(columns=['Unnamed: 0'], inplace=True)
#patients.drop(columns=['ICUSTAY_ID'], inplace=True)

data = pd.merge(data, patients, on=['subject_id','hadm_id','stay_id'], how='left')
#data = data[data['los']>1]
data.reset_index(drop=True, inplace=True)
```

```python
print(data.shape)
data.isnull().sum()
```

```python
condition3 = (data['intime'] <= data['charttime'])
condition4 = (data['outtime'] >= data['charttime'])

after_icu_data = data[condition3 & condition4]
after_icu_data = after_icu_data.sort_values(by=['subject_id','hadm_id','stay_id', 'charttime'], ascending=[True,True,True,True])
after_icu_data.reset_index(drop=True, inplace=True)
```

```python
data = 0
for bidl in bidl_ls:
    sdata = after_icu_data[after_icu_data['hadm_id']==bidl]
    if len(sdata)!=0:
        sdata_index = min(sdata.index)
        inputdata = before_icu_data_latest[before_icu_data_latest['hadm_id']==bidl].reset_index(drop=True)

        sliced_inputdata = inputdata.loc[0, 'SpO2':'Lactate']
        after_icu_data.loc[sdata_index, 'SpO2':'Lactate'] = sliced_inputdata
```

```python
after_icu_data['charttime'] = pd.to_datetime(after_icu_data['charttime'])
after_icu_data['outtime'] = pd.to_datetime(after_icu_data['outtime'])
after_icu_data['intime'] = pd.to_datetime(after_icu_data['intime'])
```

```python
def resampling_func(x):
        column = ['subject_id','hadm_id','stay_id', 'time'] + feature_name[:14]

        feature = feature_name[:14]

        x.reset_index(drop=True, inplace=True)
        date_range = pd.date_range(x['intime'][0], x['outtime'][0], freq='10T', inclusive='right')
        new_df = pd.DataFrame(index=date_range, columns=column)
        
        new_df['subject_id'] = x['subject_id'][0]
        new_df['hadm_id'] = x['hadm_id'][0]
        new_df['stay_id'] = x['stay_id'][0]
        new_df['time'] = new_df.index
        
        for i in range(len(date_range)):
                if i == 0:
                    condition1 = date_range[i]>=x['charttime']
                    sa = x[condition1]
                    for p in feature:
                        new_df[p][i] = sa[p].mean()
                else :
                    condition2 = date_range[i-1]<x['charttime']
                    condition3 = date_range[i]>=x['charttime']
                    sa = x[(condition2)&(condition3)]

                    for p in feature:
                        new_df[p][i] = sa[p].mean()
        return new_df
```

```python
result = after_icu_data.groupby('stay_id').apply(resampling_func)
result = result.reset_index(drop=True)
print('---------------------------------------------------------------------------------')
result.to_csv('./10min vital lab Dataset.csv', index=False)
```
